#### #1 Analysis of Major Trends in Climate Change over the Past Decades
- Question: What are the major trends in climate change over the past decades?
- Can be visualized using a line chart to see the trend of temperature changes and CO2 emissions over time.

```python
plt.figure(figsize=(8, 6))

# Plot average temperature trend
sns.lineplot(x='Year', y='Avg Temperature', data=data, label='Average Temperature (°C)')

# Plot per capita CO2 emissions trend
sns.lineplot(x='Year', y='CO2 Emissions', data=data, label='CO2 Emissions (Tons/Capita)')

plt.title('Climate Change Trends Over the Last Decades', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend()
plt.show()
```

From the analysis and visualization conducted on the main factors of climate change, namely average temperature and CO2 emissions, insights were obtained:
- Global Average Temperature Trend:
    - There is an increasing trend in global average temperature over the past few decades. Although there are slight fluctuations, the trend line shows a steady increase over time.
    - The increase indicates continued global warming, where average temperatures tend to increase over time, with some periods of the year experiencing higher than average peaks.
- Emisi CO₂ per Capita (CO2 Emissions):
    - CO₂ emissions per capita are relatively more stable when compared to temperature trends. Although there is a slight decrease in some periods, CO₂ emissions have not shown a sharp increase over the past few decades.
    - Although global CO₂ emissions may still be high, there have been quite successful efforts to control their growth over a certain period of time, although not enough to effectively suppress temperatures.

Overall, this data shows significant climate change in the form of global warming. While CO₂ emissions remain an important factor, the data shows that they have been controlled to some extent.


#### #2 Overall Distribution
- Question: How is the distribution of CO2 emissions per capita?
- Visualization: A histogram shows the frequency of data within a range or interval, giving an idea of ​​how CO₂ emission values ​​are distributed. This helps identify general patterns, such as what the distribution of emissions is.

```python
# CO2 emission distribution analysis
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='CO2 Emissions', bins=30, kde=True, color='skyblue')
plt.title('CO2 Emissions Distribution per capita', fontsize=14)
plt.xlabel('CO2 Emissions', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

From the histogram, we can analyze the distribution of CO₂ emissions per capita and get insights:
- CO₂ emissions per capita are relatively evenly distributed in the range from 0 to about 20 tons per capita. This shows a wide range of emissions, indicating variability in how countries contribute to CO₂ emissions per person.
- The histogram appears to approach a uniform distribution with almost similar frequencies in most intervals, except for a few peaks, such as around 5 and 10 tons per capita. This may indicate that although most countries have varying levels of emissions, some levels of emissions are more common.
- The KDE (Kernel Density Estimate) curve shows mild fluctuations, with a moderate peak around 5 tons per capita and a slight increase around 15-17 tons. This implies that there may be a cluster of countries with emissions concentrated around this level.
- Unlike a normal distribution that has a single central peak, this distribution is relatively flat, indicating that there is no single level of CO₂ emissions per capita that can be considered typical. Instead, emissions levels are spread across a wide range.

Thus, CO₂ emissions per capita vary widely across countries, with no single level of emissions that can be considered typical. This may reflect the diversity in industrial activity, energy sources, and environmental policies across countries.

#### #3 Analysis of Global Temperature Mean and Distribution
- Question: What is the global temperature mean in the dataset and how is it distributed?
- Purpose: Provide a descriptive summary of global temperature to understand the characteristics of the data
- Analysis method: Descriptive statistics such as mean, median, standard deviation, and temperature distribution with histogram.

```python
# Descriptive statistics of global average temperature
country_temp_stats = data["Avg Temperature"].describe()

print(country_temp_stats)

# Get Q1, Q3, and IQR
Q1 = country_temp_stats["25%"]
Q3 = country_temp_stats["75%"]
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = data[(data["Avg Temperature"] < lower_bound) | (data["Avg Temperature"] > upper_bound)]

# Count the number of outliers
outlier_count = len(outliers)

# Display the number of outliers
print(f"Number of outliers: {outlier_count}")
```

From the results of statistical analysis of the average global temperature, insights were obtained:
- The average global temperature is 19.88°C, with a standard deviation of 8.54°C. This shows a fairly large variation in temperature throughout the world.
- The minimum temperature is 5.00°C, while the maximum temperature reaches 34.90°C, which means the temperature range is quite wide.
- The median temperature (20.10°C) which is close to the average shows a fairly symmetrical temperature distribution without the dominance of extreme data.
- No outliers were found based on IQR, so the data distribution is most likely normal or close to normal and shows that the dataset does not have extreme values ​​that can affect further analysis.
- The data range and quartile distribution show temperature variations that represent global climate diversity, from low to high temperature locations.

```python
# Histogram plot for global temperature distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="Avg Temperature", bins=30, kde=True, color="skyblue")
plt.title("Global Mean Temperature Distribution", fontsize=16)
plt.xlabel("Average Temperature (°C)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

In the histogram above we can say that:
- Histogram shows the distribution of global average temperatures with a temperature range of around 5°C to 35°C.
- The distribution looks relatively even, with a few peaks in the ranges of 10°C - 15°C and 25°C - 30°C.
- The increase in the frequency of certain temperatures can indicate areas with specific climate conditions (for example, more stable subtropics).
- The curved line above the histogram indicates the possibility of a normal distribution, although the data has a wide spread.

#### #4 Correlation Analysis: Sea Level Rise, Precipitation, and Extreme Events

- Question: Why might sea level rise cause changes in precipitation and extreme events?
- Visualization: Heatmap to show the correlation matrix between the variables
- Analysis: Calculate the correlation matrix using the corr() function in Pandas

```python
corr_matrix = data[['Sea Rise', 'Rainfall', 'Extreme Events']].corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

Based on the graph, it can be concluded that:

- There is no strong correlation between the three variables. The correlation coefficient value between the three variables is very small and approaches zero. This means that changes in one variable do not significantly affect changes in other variables.
- Sea Level Rise has a very weak positive correlation with Rainfall and Extreme Events. This indicates that when Sea Level Rise increases, Rainfall and Extreme Events also tend to increase slightly, but this relationship is very weak and not statistically significant.
- Rainfall has a very weak negative correlation with Extreme Events. This shows that when Rainfall increases, the number of Extreme Events tends to decrease slightly, but this relationship is also very weak and not statistically significant.

#### #6 Extreme Weather Events Trend Analysis
- Question: Are there any specific years that show a significant spike or drop in the number of extreme weather events?
- Analysis Method: Line graph to visualize extreme weather events per year with distribution analysis.

```python
# Filter data for available year range (2000 to 2023)
data_recent_23_years = data[(data['Year'] >= 2000) & (data['Year'] <= 2023)]

# Aggregate number of extreme weather events per year
extreme_events_per_year = data_recent_23_years.groupby('Year')['Extreme Events'].sum().reset_index()

# Create a line plot for extreme weather event trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=extreme_events_per_year, x='Year', y='Extreme Events', marker='o', color='b')
plt.title("Extreme Weather Event Trends (2000-2023)")
plt.xlabel("Year")
plt.ylabel("Number of Extreme Weather Events")

# Highlights of the last decade (2013 - 2023)
plt.axvspan(2013, 2023, color="orange", alpha=0.3, label="Last Decade")
plt.legend()

plt.show()
```

The visualization results show that although the number of extreme weather events overall decreased from 2000 to 2010, the last decade shows a large fluctuation pattern with an unstable up-and-down pattern that may be of concern to policy makers.
- A significant spike in extreme weather events occurred in 2017 (where the number of extreme weather events increased sharply compared to the previous year) and 2020 (the highest number of extreme weather events in the last decade)
- A significant decrease in extreme weather: Occurred in 2016, when the number of extreme weather events decreased sharply compared to the previous year (2015).

#### #7 Analysis of Variation in Renewable Energy Use Levels Between Countries

- Question: How is the variation in renewable energy use? Which countries have the highest and lowest renewable energy use based on the variation?
- Visualization Method: Boxplot
- Analysis Steps:
    - Calculate the mean, standard deviation, and variance of the Renewable Energy column.
    - Use a boxplot to depict the distribution of the data, including outliers, medians, and quartiles.
    - Identify countries with high and low renewable energy use.

```python
# Statistical Analysis
mean_renewable = data['Renewable Energy'].mean()
std_renewable = data['Renewable Energy'].std()
var_renewable = data['Renewable Energy'].var()

print(f"Mean: {mean_renewable:.2f}")
print(f"Standard Deviation: {std_renewable:.2f}")
print(f"Variance: {var_renewable:.2f}")

# Boxplot visualization
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, y='Country', x='Renewable Energy', color='skyblue')
plt.title('Renewable Energy Variations between Countries', fontsize=14)
plt.ylabel('Country', fontsize=12)
plt.xlabel('Renewable Energy (%)', fontsize=12)
plt.show()
```

Based on this boxplot, several analyses can be drawn regarding the distribution of renewable energy use across countries:

1. Countries with High Renewable Energy Use:
    - Mexico has a higher median, the center line on the boxplot box is at 30%, indicating that half of its usage data is above that value. And Mexico also has a wide interquartile range.
    - China also has a high median reaching more than 30%. So that shows that China and Mexico are the countries with the highest use of renewable energy.

2. Countries with Low Renewable Energy Use:
    - South Africa has a small quartile range compared to the others and the lowest median (slightly above 20%). So it can be said that from the data recorded in the dataset, South Africa is a country that uses little renewable energy.

3. Distribution:
    - Countries such as France and Germany show a fairly wide range, indicating significant variation between their uses. This could be influenced by regional policies or local resources.
    - On the other hand, Japan has a narrower range, indicating stability in the proportion of renewable energy in the country.

4. Countries with Outliers or Anomalies:
    - There are no obvious outliers in this boxplot. All values ​​appear to be within a reasonable range, although there is some variation between countries.

#### #8 Analysis of Forest Area Changes Over Time
- Question: How is the cumulative change in forest area over time?
- Visualization Method: Area plots are useful for displaying the accumulated changes in forest area over time. This will give an idea of ​​how the forest area is evolving and show the cumulative changes more clearly.

```python
plt.figure(figsize=(10, 6))
data.groupby('Year')['Forest Area'].mean().plot(kind='area', color='lightgreen', alpha=0.5)
plt.title('Cumulative Forest Area Over Time')
plt.xlabel('Year')
plt.ylabel('Forest Area (%)')
plt.grid(True)
plt.show()
```

From the Area plot above, it can be detailed that:
- Forest Area Fluctuation: The graph shows that the forest area experienced quite significant fluctuations during the observed period. There were periods where the forest area increased, but there were also periods where the forest area decreased.
- In general, the graph shows a fairly stable trend. Despite the fluctuations, the overall forest area did not experience a drastic decline.
- Peak and Lowest Points: There are several points where the forest area reaches its highest and lowest peaks. The peak indicates a period where conservation and reforestation efforts succeeded in increasing the forest area, while the lowest point indicates a period where the deforestation rate was higher.
- After reaching a peak around 2010, the forest area tended to be stable or experienced a slight decline. This could indicate that the conservation efforts carried out have provided quite good results in maintaining the forest area.

#### #9 Projections of global average temperature changes in the next 10 years

- Question: Based on current trends, what is the projection of global average temperature changes in the next 10 years with a minimum accuracy of 85%?
- Analysis Method: Inferential statistical analysis using confidence intervals or time series regression for temperature increase projections.
- Visualization: Line plot of average temperature trends and their projections.

Process:
- The initial data is processed by calculating the average global temperature per year grouped by year to produce an annual trend and then can be used in the analysis
- The average temperature trend is visualized with a line plot to see the historical pattern of how temperature changes from year to year
- Linear regression for temperature projection:
    - using Year and Avg Temperature as variables.
    - Then a constant is added to x so that the model can take the intercept in the linear regression.
    - then create a simple linear regression model with Ordinary Least Squares and display the data for easy analysis.
- Project the temperature 10 years into the future: where the next year is calculated based on the current trend.
- Calculate the Confidence Interval (CI) for the projection where this CI takes into account model uncertainty with a 95% CI to provide a more realistic range of possible temperatures.
- Then visualize the projection that combines historical trends, projections, and uncertainty.
- After that, the model is evaluated by calculating the MAE (Mean Absolute Error) and model accuracy.
